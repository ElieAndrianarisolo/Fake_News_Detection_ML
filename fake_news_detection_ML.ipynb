# Fake News Detection using Machine Learning

```python
import pandas as pd
import numpy as np
```

### Reading CSV files

```python
true = pd.read_csv('True.csv')
fake = pd.read_csv('Fake.csv')
```

```python
true.head()
```

```python
fake.head()
```

### Combining the 2 Dataframes

```python
true['label'] = 1
fake['label'] = 0
```

```python
true.head()
```

```python
fake.head()
```

```python
news = pd.concat([fake, true], axis = 0)
```

```python
news.head()
```

```python
news.tail()
```

```python
news.isnull().sum()
```

```python
news = news.drop(['title', 'subject', 'date'], axis = 1)
```

```python
news.head()
```

###Shuffling the data to avoid any bias

```python
news = news.sample(frac = 1)
```

```python
news.head()
```

```python
news.reset_index(inplace=True)
```

```python
news.head()
```

```python
news.drop(['index', 'level_0'], axis=1, inplace=True)
```

```python
news.head()
```

### Cleaning the data

```python
import re
```

```python
def wordopt(text):

  # Convert into lowercase
  text = text.lower()

  # Remove URLs
  text = re.sub(r'https?://\S+|www\.\S+', '', text)

  # Remove HTML tags
  text = re.sub(r'<.*?>', '', text)

  # Remove punctuation
  text = re.sub(r'[^\w\s]', '', text)

  # Remove digits
  text = re.sub(r'\d', '', text)

  # Remove newline characers
  text = re.sub(r'\n', ' ', text)

  return text
```

```python
news['text'] = news['text'].apply(wordopt)
```

```python
news['text']
```

### Logistic regression

```python
X = news['text']
Y = news['label']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
Xv_train = vectorization.fit_transform(X_train)
Xv_test = vectorization.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

LR = LogisticRegression()
LR.fit(Xv_train, Y_train)

pred_lr = LR.predict(Xv_test)
```

```python
LR.score(Xv_test, Y_test)
```

```python
print(classification_report(Y_test, pred_lr))
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

DTC = DecisionTreeClassifier()
DTC.fit(Xv_train, Y_train)

pred_dtc = DTC.predict(Xv_test)
```

```python
DTC.score(Xv_test, Y_test)
```

```python
print(classification_report(Y_test, pred_dtc))
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

RFC = RandomForestClassifier()
RFC.fit(Xv_train, Y_train)

pred_rfc = RFC.predict(Xv_test)
```

```python
RFC.score(Xv_test, Y_test)
```

```python
print(classification_report(Y_test, pred_rfc))
```

### Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier

GBC = GradientBoostingClassifier()
GBC.fit(Xv_train, Y_train)

pred_gbc = GBC.predict(Xv_test)
```

```python
GBC.score(Xv_test, Y_test)
```

```python
print(classification_report(Y_test, pred_gbc))
```

### Manuel Testing of the predictive model

```python
def output_label(n):

  if n == 0:
    return "Fake news"
  elif n == 1:
    return "Genuine news"
```

```python
def manual_testing(news):

  testing_news = {"text" : [news]} # Corrected syntax for defining dictionnary
  new_def_test = pd.DataFrame(testing_news)
  new_def_test["text"] = new_def_test["text"].apply(wordopt)

  new_x_test = new_def_test["text"]
  new_xv_test = vectorization.transform(new_x_test)

  pred_lr = LR.predict(new_xv_test)
  pred_dtc = DTC.predict(new_xv_test)
  pred_gbc = GBC.predict(new_xv_test)
  pred_rfc = RFC.predict(new_xv_test)

  return "\n\nLR Prediction {} \nDTC Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(output_label(pred_lr[0]), output_label(pred_dtc[0]), output_label(pred_gbc[0]), output_label(pred_rfc[0]))
```

```python
news_article = str(input())
```

```python
print(manual_testing(news_article))
```
